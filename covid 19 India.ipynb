# covid-19 India case study
1. import covid 19 data
2. Keeping only required columns (Date, State, Cured, Death, Confirmed)
3. Dig a data for a specific date
4. Arrange the confirmed data into ascending order
5. Analyse any one of the state for cured, death and confirmed cases

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
import datetime as dt
```

```python
covid = pd.read_csv("D:\CSV\covid_19_india.csv", parse_dates=['Date'])   #to convert the string date into date format
covid.head()
```

```python
covid = covid[['Date', 'State/UnionTerritory', 'Cured', 'Deaths', 'Confirmed']]    #getting only the required columns
covid.columns = ['date', 'state', 'cured', 'deaths', 'confirmed']    #changing the name of columns for convenience or any other reason
```

```python
covid.head()
```

```python
now = covid[covid.date == '2020-07-17']   #to extract the data for the particular date
now.head()
```

```python
#to put the values of particular date into ascending order
maxcase=now.sort_values(by='confirmed', ascending=False)
maxcase1=maxcase[0:5]   #sorting the max covid case in top 5 states
```

```python
maxcase1
```

```python
sns.barplot(x="state",y="confirmed", data=maxcase1, hue="state")
plt.show()
```

```python
maxdeath = now.sort_values(by='deaths', ascending=False)
maxdeath1=maxdeath[0:5]  #max deaths in top 5 states
```

```python
maxdeath1
```

```python
sns.barplot(x='state', y='deaths', data=maxcase1)
plt.show()
```

```python
cured = now.sort_values(by="cured", ascending=False)
cured1 = cured[0:5]
```

```python
cured1
```

```python
sns.barplot(x="state", y="cured", data=cured1, hue="state")
plt.show()
```
